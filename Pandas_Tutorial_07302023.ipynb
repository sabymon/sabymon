# Dataframe and Series Basics

Basic pd data types dataframe and series. df in pd can be a dictionary in python

```python
people = {
    "A": ["abba", "dabba", "jabba"],
    "B": ["marko", "tula", "tampa"]
}
```

```python
people
```

```python
people['B']
```

Create a dataframe from dictionary

```python
df_dict = pd.DataFrame (people)
df_dict
```

Read from data frame

```python
df_dict['B']
```

Note that > people['B'] and > df_dict ['B'] returning different values. Why? df_dict [] is returning a series, in other case a dictionary of list.
